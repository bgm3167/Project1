```python
stock = pd.read_csv(r"C:\Users\rhksa\OneDrive\바탕 화면\sesac-first project\stock.csv")
```

```python
stock['날짜'] = stock['날짜'].str.replace('.', '', regex=False).str.strip()
```

```python
stock
```

```python
stock = stock.sort_values(by='날짜', ascending=True)
stock = stock.reset_index(drop=True)
```

```python
stock
```

```python
stock['날짜'] = stock['날짜'].astype(int)
answer = pd.merge(c_Samsung, stock, on='날짜', how='inner')
```

```python
answer
```

```python
answer['삼성전자주가'] = answer['날짜'].astype(int)
answer['본문%'] = answer['데이터 수']/answer['데이터 수'].sum()*100
answer['삼성전자주가%'] = answer['삼성전자주가']/answer['삼성전자주가'].sum()*100
```
